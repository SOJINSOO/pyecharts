### - Load Packages 

```python
import random
from pyecharts import Bar
```

### - Data Setting

```python
CLOTHES = ["셔츠", "울셔츠", "블라우스", "바지", "하이힐", "양말"]

clothes_v1 = [5, 20, 36, 10, 75, 90]
clothes_v2 = [10, 25, 8, 60, 20, 80]
```

### 1. bar_stack

```python
bar = Bar()
bar.add("가게A", CLOTHES, clothes_v1, is_stack=True)
bar.add("가게B", CLOTHES, clothes_v2, is_stack=True)
bar.chart_id = "id_chart_id"
html_content = bar._repr_html_()
assert "id_chart_id" == bar.chart_id
assert "dataZoom" not in html_content
assert "stack_" in html_content

bar
```

### 2. bar_marks

```python
bar = Bar("mark_point 및 mark_line 예시")
bar.add("가게A", CLOTHES, clothes_v1, mark_point=["average"])
bar.add("가게B", CLOTHES, clothes_v2, mark_line=["min", "max"])
assert '"average"' in bar._repr_html_()

bar
```

### 3. bar_convert_xy_axis

```python
bar = Bar("x 축과 y 축 전환")
bar.add("가게A", CLOTHES, clothes_v1)
bar.add("가게B", CLOTHES, clothes_v2, is_convert=True)
assert "average" not in bar._repr_html_()

bar
```

### 4. bar_histogram

```python
bar = Bar("Histogram 예시")
bar.add("", CLOTHES * 2, clothes_v1 + clothes_v2, bar_category_gap=0)
bar
```

### 5. bar_rotate_label

```python
days = ["{}일".format(i) for i in range(20)]
days_v1 = [random.randint(1, 20) for _ in range(20)]
bar = Bar("축 레이블 회전 예시")
bar.add("", days, days_v1, xaxis_interval=0, xaxis_rotate=30, yaxis_rotate=30)
assert "stack_" not in bar._repr_html_()

bar
```

### 6. bar_waterfall

```python
months = ["{}월".format(i) for i in range(1, 8)]
months_v1 = [0, 100, 200, 300, 400, 220, 250]
months_v2 = [1000, 800, 600, 500, 450, 400, 300]
bar = Bar("Waterfall 예시")
bar.add("", months, months_v1, label_color=["rgba(0,0,0,0)"], is_stack=True)
bar.add(
    "월",
    months,
    months_v2,
    is_label_show=True,
    is_stack=True,
    label_pos="inside",
)

bar
```

### 7. bar_datazoom_undefined

```python
days = ["{}일".format(i) for i in range(30)]
days_v1 = [random.randint(1, 30) for _ in range(30)]
bar = Bar("Bar - datazoom 묵시적 예시")
bar.add("", days, days_v1, is_label_show=True, is_datazoom_show=True)
html_content = bar._repr_html_()
assert "dataZoom" in html_content
assert ': "slider"' in html_content
assert ': "inside"' not in html_content

bar
```

### 8. bar_datazoom_slider

```python
days = ["{}일".format(i) for i in range(30)]
days_v1 = [random.randint(1, 30) for _ in range(30)]
bar = Bar("Bar - datazoom 예시")
bar.add(
    "",
    days,
    days_v1,
    is_datazoom_show=True,
    datazoom_type="slider",
    datazoom_range=[10, 25],
)
html_content = bar._repr_html_()
assert "dataZoom" in html_content
assert ': "slider"' in html_content
assert ': "inside"' not in html_content

bar
```

### 9. bar_datazoom_inside

```python
days = ["{}일".format(i) for i in range(30)]
days_v1 = [random.randint(1, 30) for _ in range(30)]
bar = Bar("Bar - datazoom - inside 예시")
bar.add(
    "",
    days,
    days_v1,
    is_datazoom_show=True,
    datazoom_type="inside",
    datazoom_range=[10, 25],
)
html_content = bar._repr_html_()
assert "dataZoom" in html_content
assert ': "inside"' in html_content
assert ': "slider"' not in html_content

bar
```

### 10. bar_datazoom_both

```python
days = ["{}일".format(i) for i in range(30)]
days_v1 = [random.randint(1, 30) for _ in range(30)]
bar = Bar("Bar - datazoom - both 예시")
bar.add(
     "",
    days,
    days_v1,
    is_datazoom_show=True,
    datazoom_type="both",
    datazoom_range=[10, 25],
    is_toolbox_show=False,
)
assert len(bar.options.get("dataZoom")) == 2
html_content = bar._repr_html_()
assert "dataZoom" in html_content
assert ': "inside"' in html_content
assert ': "slider"' in html_content

bar
```

### 11. test_bar_datazoom_xaxis_and_yaxis

```python
days = ["{}일".format(i) for i in range(30)]
days_v1 = [random.randint(1, 30) for _ in range(30)]
bar = Bar("Bar - datazoom - xaxis/yaxis 예시")
bar.add(
    "",
    days,
    days_v1,
    # 묵시적 X 축, 가로 방향
    is_datazoom_show=True,
    datazoom_type="slider",
    datazoom_range=[10, 25],
    # 추가적인 dataZoom 제어 막대 추가, 세로 방향
    is_datazoom_extra_show=True,
    datazoom_extra_type="slider",
    datazoom_extra_range=[10, 25],
    is_toolbox_show=False,
)
assert len(bar.options.get("dataZoom")) == 2
html_content = bar._repr_html_()
assert "dataZoom" in html_content
assert ': "slider"' in html_content
assert ': "vertical"' in html_content
assert ': "horizontal"' in html_content

bar
```

### 12. test_bar_extra_html_text_label

```python
bar = Bar("Column chart", extra_html_text_label=["bar_extra_html_text_label"])
bar.add("가게A", CLOTHES, clothes_v1, is_stack=True)
bar.add("가게B", CLOTHES, clothes_v2, is_stack=True)
html_content = bar._repr_html_()
assert '<p style="">bar_extra_html_text_label</p>' in html_content

bar
```

### 13. test_bar_line_color_and_with

```python
bar = Bar("Column chart")
bar.add(
    "가게A",
    CLOTHES,
    clothes_v1,
    xaxis_line_color="green",
    xaxis_line_width=5,
    xaxis_label_textcolor="black"
)
html_content = bar._repr_html_()
assert '"color": "green"' in html_content
assert '"width": 5' in html_content

bar
```
