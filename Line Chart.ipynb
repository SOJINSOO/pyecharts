# 2. Line

### - Load Packages

```python
import math
import random

from pyecharts import Line
```

### - Data Setting

```python
CLOTHES = ["셔츠", "스웨터", "블라우스", "바지", "하이힐", "양말"]
WEEK = ["월요일", "화요일", "수요일", "목요일", "금요일", "토요일", "일요일"]

clothes_v1 = [5, 20, 36, 10, 10, 100]
clothes_v2 = [55, 60, 16, 20, 15, 80]
```

### 1. line user define markpoint

```python
line = Line("Line Chart 예제")
line.add(
    "가게A",
    CLOTHES,
    clothes_v1,
    mark_point=["average", {"coord": ["바지", 10], "name" : "First Point"}]
)
line.add(
    "가게B",
    CLOTHES,
    clothes_v2,
    is_smooth= True,
    mark_point=[{"coord": ["양말", 80], "name": "Second Point"}]
)

line
```

### 2. line user define marks

```python
line = Line("Line Chart 예제")
line.add(
    "가게A",
    CLOTHES,
    clothes_v1,
    mark_point=["average", "max", "min"],
    symbol_size=50,
    mark_point_symbol="diamond",
    mark_point_textcolor="#40ff27",
)
line.add(
    "가게B",
    CLOTHES,
    clothes_v2,
    mark_point=["average", "max", "min"],
    mark_point_symbol="arrow",
    mark_point_symbolsize=40,
)

line
```

### 3. line type step

```python
line = Line("꺾은 선형 차트 - 단계 다이어그램 예제")
line.add("가계A", CLOTHES, clothes_v1, is_step=True, is_label_show=True)

line
```

### 4. line type fill

```python
line = Line("꺾은 선형 차트 - 영역 차트 예제")
line.add(
    "가게A",
    CLOTHES,
    clothes_v1,
    is_fill=True,
    line_opacity=0.2,
    area_opacity=0.4,
    symbol=None,
)
line.add(
    "가게B",
    CLOTHES,
    clothes_v2,
    is_fill=True,
    area_color="#000",
    area_opacity=0.3,
    is_smooth=True,
)

line
```

### 5. line mark point raw

```python
line = Line()
line.add(
    "가게A",
    CLOTHES,
    clothes_v1,
    mark_point_raw=[
        {"name": "rawData", "symbol": "pin", "coord": ["셔츠", 5], "value": 5}
    ]
)

line
```
