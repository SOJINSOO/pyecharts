# 3. Pie

### - Load Packages

```python
import random
from pyecharts import Pie, Style
```

### - Data Setting

```python
CLOTHES = ["셔츠", "스웨터", "블라우스", "바지", "하이힐", "양말"]
```

### 1. pie default formatter

```python
v1 = [11, 12, 13, 10, 10, 10] 
pie = Pie("Pie Chart 예제") 
pie.add("", CLOTHES, v1, is_label_show=True) 

pie
```

### 2. pie legend

```python
v1 = [11, 12, 13, 10, 10, 10]
pie = Pie("Pie Chart - Ring Chart 예제", title_pos="center")
pie.add(
    "",
    CLOTHES,
    v1,
    radius=[40, 75],
    label_text_color=None,
    is_label_show=True,
    legend_orient="vertical",
    legend_pos="left",
)

pie
```

### 3. pie type rose

```python
v1 = [11, 12, 13, 10, 10, 10]
v2 = [19, 21, 32, 20, 20, 33]
pie = Pie("Pie Chart - Rose Figure 예제", title_pos="center", width=900)
pie.add(
    "상품A",
    CLOTHES,
    v1,
    center=[25, 50],
    is_random=True,
    radius=[30, 75],
    rosetype="radius",
)
pie.add(
    "상품B",
    CLOTHES,
    v2,
    center=[75, 50],
    is_random=True,
    radius=[30, 75],
    rosetype="area",
    is_legend_show=False,
    is_label_show=True,
)

pie
```

### 4. pie type radius

```python
pie = Pie("Pie Chart 예제", title_pos="center", width=1000, height=600)
pie.add(
    "",
    ["A", "B", "C", "D", "E", "F"],
    [335, 321, 234, 135, 251, 148],
    radius=[40, 55],
    is_label_show=True,
)
pie.add(
    "",
    ["H", "I", "J"],
    [335, 679, 204],
    radius=[0, 30],
    legend_orient="vertical",
    legend_pos="left",
)

pie
```

### 5. pie multiple

```python
attr = ["A", "B", "C", "D", "E", "F"]
pie = Pie("Pie Chart 예제", width=1000, height=600)
pie.add(
    "",
    attr,
    [random.randint(0, 100) for _ in range(6)],
    radius=[50, 55],
    center=[25, 50],
    is_random=True,
)
pie.add(
    "",
    attr,
    [random.randint(20, 100) for _ in range(6)],
    radius=[0, 45],
    center=[25, 50],
    rosetype="area",
)
pie.add(
    "",
    attr,
    [random.randint(0, 100) for _ in range(6)],
    radius=[50, 55],
    center=[65, 50],
    is_random=True,
)
pie.add(
    "",
    attr,
    [random.randint(20, 100) for _ in range(6)],
    radius=[0, 45],
    center=[65, 50],
    rosetype="radius",
)

pie
```

### 6. pie multiple movie 

```python
pie = Pie("영화 장르 중 '좋아하는 장르' 비율", title_pos="center")

style = Style()
pie_style = style.add(label_pos="center", is_label_show=True, label_text_color=None)

pie.add("", ["Plot", ""], [25, 75], center=[10, 30], radius=[18, 24], **pie_style)
pie.add("", ["Fantasy", ""], [24, 76], center=[30, 30], radius=[18, 24], **pie_style)
pie.add("", ["Love", ""], [14, 86], center=[50, 30], radius=[18, 24], **pie_style)
pie.add("", ["Horror", ""], [11, 89], center=[70, 30], radius=[18, 24], **pie_style)
pie.add("", ["Adventure", ""], [27, 73], center=[90, 30], radius=[18, 24], **pie_style)
pie.add("", ["Action", ""], [15, 85], center=[10, 70], radius=[18, 24], **pie_style)
pie.add("", ["Comedy", ""], [54, 46], center=[30, 70], radius=[18, 24], **pie_style)
pie.add("", ["Science fiction", ""], [26, 74], center=[50, 70], radius=[18, 24], **pie_style)
pie.add("", ["Suspense", ""], [25, 75], center=[70, 70], radius=[18, 24], **pie_style)
pie.add(
    "",
    ["Crime", ""],
    [28, 72],
    center=[90, 70],
    radius=[18, 24],
    legend_top="center",
    **pie_style
)

pie
```
